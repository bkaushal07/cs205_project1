```python
import matplotlib.pyplot as plt
```

```python
depths = [0, 2, 4, 8, 12, 16, 20, 24]

uniform_cost_search_times = [0, 0.003, 0.013, 0.085, 1.315, 64.449, 1541.797, 19852.429]
misplaced_tile_times = [0, 0.001, 0.003, 0.007, 0.044, 0.276, 3.656, 157.595]
manhattan_distance_times = [0, 0.002, 0.003, 0.005, 0.012, 0.038, 0.191, 1.142]

plt.plot(depths, uniform_cost_search_times, marker='o', label='Uniform Cost Search')
plt.plot(depths, misplaced_tile_times, marker='o', label='A* with Misplaced Tile')
plt.plot(depths, manhattan_distance_times, marker='o', label='A* with Manhattan Distance')

plt.xticks(depths)
plt.xlabel('Depth of Puzzle')
plt.ylabel('Time (seconds)')
plt.title('Time vs Depth of Puzzle for Different Search Algorithms')
plt.legend()

plt.show()
```
